## 1.2 Create a three-dimensional vecor and plot it on a graph

```
from mpl_toolkits.mplot3d import Axes3D

# Create 3d vector
b = [3,6,9]

# Create origin vector
origin = [0,0,0]

# zip so we have 2 points for each axis
X, Y, Z = zip(origin, b)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot vector
ax.quiver(X[0], Y[0], Z[0], X[1], Y[1], Z[1], color='k')

ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.set_zlim(0, 10)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.title('3D Vector (3,6,9)');
```

## 2.2 Find the following products: CD, AE, and BA. What are the dimensions of the resulting matrices? How does that relate to the dimensions of their factor matrices?

```
# Create the matrices
A = np.array([[1, 2],
              [3, 4],
              [5, 6]])

B = np.array([2, 4, 6])

C = np.array([[9, 6, 3],
              [4, 7, 11]])

D = np.array([[1, 0, 0],
              [0, 1, 0],
              [0, 0, 1]])

E = np.array([[1, 3],
              [5, 7]])

# Check that they came out as planned
A, B, C, D, E
```

```
# Calculate and print out the matrices multiplied
print('CD = ', np.matmul(C, D))
print('AE = ', np.matmul(A, E))
print('BA = ', np.matmul(B, A))
```

CD is (2, 3) the same dimensions as C was before (its also the exact same matrix bec it was multiplied by the identity matrix)

AE is (3, 2) the same shape as A was before

BA is (2,) not the shape of either matrix before

The matrix that is the product of matrix multiplication has the lowest denominator dimensions of the 2 multiplied matrices (i.e. it has the rows of the one with fewer rows and the columns of the fewer columns)

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
# Create the matrix
H = np.array([[12, 11],
              [7, 10]])

# Calculate the determinant
np.linalg.det(H)
```

```
# Create the matrix
J = np.array([[0, 1, 2],
              [7, 10, 4],
              [3, 2, 0]])

# Calculate the determinant
np.linalg.det(J)
```

## 3.3 Find $H^{-1}$ and then find $J^{-1}$

```
# Calculate inverse of H and store to variable
H_inv = np.linalg.inv(H)
H_inv
```

```
# Calculate inverse of J and store to variable
J_inv = np.linalg.inv(J)
J_inv
```

## 3.4 Find $HH^{-1}$ and then find $J^{-1}J$. Is $HH^{-1} == J^{-1}J$? Why or Why not?

```
# Calculate and print the matrices multiplied by their inverse
print('HH`1 =', np.matmul(H, H_inv))
print('J`1J =', np.matmul(J_inv, J))
```

They are not equivalent but they are both the identity matrix. They are both identity matrices because a matrix multiplied by its inverse equals the identity matrix, but they are not the same bec they have different dimensions, due to the original matrices being of differing dimension
